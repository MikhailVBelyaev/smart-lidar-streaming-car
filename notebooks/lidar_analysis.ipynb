# LiDAR Data Exploration
This notebook loads Parquet files generated by Spark Structured Streaming and performs basic analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_parquet('/tmp/lidar_output')
df.head()
```

```python
# Plot the LiDAR scan
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['angle'], y=df['distance'])
plt.title('LiDAR Angle vs Distance')
plt.xlabel('Angle (degrees)')
plt.ylabel('Distance (meters)')
plt.grid(True)
plt.show()
```
